# Shuffling the data for robustness of the ML model

```python
data_tuples=np.array(data)
```

```python
data_tuples
```

```python
import random
random.shuffle(data_tuples)
```

```python
X = [i[0] for i in data_tuples]
```

```python
X
```

```python
y = [i[1] for i in data_tuples]
```

```python
y
```

# Dividing the list into characters

```python
def dividing(input):
    characters=[]
    for i in input:
        characters.append(i)
    return characters
```

```python
dividing(X[0])
```

# Tf-idf for vectorizing the characters

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(tokenizer=dividing)
```

```python
x = vectorizer.fit_transform(X)
```

```python
x
```

```python
x.shape
```

```python
vectorizer.get_feature_names_out()
```

```python
first_document_vector =x[0]
```

```python
first_document_vector
```

```python
first_document_vector.T.todense()
```

```python
df=pd.DataFrame(first_document_vector.T.todense(),index=vectorizer.get_feature_names_out(),columns=['Tf-idf'])
```

```python
df.sort_values('Tf-idf',ascending=False)
```

# Applying ML algorithm Logistic Regression

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2) #20% will be used as testing data
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classification = LogisticRegression(random_state=0,multi_class='multinomial').fit(X_train,y_train) #multinomial as there are three categories 0,1,2
```

# Checking a random password strength

```python
password = np.array(['awsd45gff^&@#d'])
pred = vectorizer.transform(password)
classification.predict(pred)
```

```python
y_pred = classification.predict(X_test)
```

```python
y_pred
```

# Checking the accuracy of the model

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm = confusion_matrix(y_test,y_pred)
```

```python
print(cm)
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
